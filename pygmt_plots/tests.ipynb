```python
import pygmt
import os
import geopandas as gpd
import matplotlib as plt
import matplotlib_inline
path = os.getcwd().split('/')
machine_path = '/'+path[1]+'/'+path[2]
# pygmt.show_versions()

fig = pygmt.Figure()

region = [-180, 180, -90, 90]
# projection = 'Cyl_stere/30/-20/12c'
# projection = 'W12c'
projection = 'Kf12c' 
# projection = 'N12c'
# projection = 

#plot coast
fig.coast(region=region,
          projection=projection,
          resolution='h',
          shorelines=True,
          frame=True, #set frame and axes attributes.
        #   water='LIGHTCYAN',
        #   water='ROYALBLUE4',
          water='ROYALBLUE3',
          land='224')

#creating basemap
fig.basemap(region=region,
            projection=projection,
            frame="a30g"
            )

#read shapefiles data
path = f"{machine_path}/opt/global_tectonics/plates&provinces/shp"
filename = f"{path}/cratons.shp"
gdf = gpd.read_file(filename)

fig.plot(filename, fill='PERU')
fig.text(text='Cratonic regions', x=-125, y=-32, font='10p,Helvetica-Bold,PERU', fill='white')
fig.show()
figname = 'cratonic_regions'

fig.savefig(f"{figname}.pdf", dpi=150)
```
